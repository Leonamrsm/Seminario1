```python
# Import all the libraries that we shall be using
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import scipy

from scipy.spatial import distance_matrix
```

Para este exercíco foram geradas duas distribuições normais caracterizadas como ($\mu_1$, $\sigma_1$) e ($\mu_2$, $\sigma_2$), com 50 pontos cada ($N_1 = N_2 = 50$). Conforme o código a seguir:

```python
mu1 = [-1, -1]
sigma1 = [[0.7, 0], [0, 0.7]]  # diagonal covariance
N1 = 50
X_1 = np.random.multivariate_normal(mu1, sigma1, N1)

mu2 = [1, 1]
sigma2 = [[0.7, 0], [0, 0.7]]  # diagonal covariance
N2 = 50
X_2 = np.random.multivariate_normal(mu2, sigma2, N2)

def plotPoints(X, Y):
    plt.scatter(X[np.where(Y == -1)[0],0], X[np.where(Y == -1)[0],1])
    plt.scatter(X[np.where(Y == 1)[0],0], X[np.where(Y == 1)[0],1])
    plt.xlabel('X1')
    plt.ylabel('X2')

#Todos os pontos
X = np.vstack((X_1, X_2))
#Classes de todos os pontos
Y = np.vstack((-1*np.ones((N1, 1)), np.ones((N2, 1))))
#Sepera as classes únicas
classes = np.unique(Y)

plotPoints(X, Y)
```

```python
def GabrielGraph(X, Y, plot):

    if(plot):
        plotPoints(X, Y)

    G = np.zeros((X.shape[0],X.shape[0]))

    D = distance_matrix(X, X)

    for i in range(X.shape[0]):
        xi = X[i,:]
        for j in range(X.shape[0]):
            xj = X[j,:]
            aux = 0
            for k in range(X.shape[0]):
                xk = X[k,:]
                if k == i or k == j:
                    continue
                if D[i][j]**2 <= (D[i][k]**2 + D[j][k]**2):
                    aux = aux + 1
            if(aux == (X.shape[0] - 2)):
                G[i,j] = 1
                x_values = [xi[0], xj[0]]
                y_values = [xi[1], xj[1]]
                if(plot):
                    plt.plot(x_values, y_values, 'k')
    return G
                
G = GabrielGraph(X, Y, True)
```

```python
def getSupportEdges(X, Y, G, plot):
    
    if(plot):
        plotPoints(X, Y)
    
    SE = []

    for i in range(X.shape[0]):
        x1 = X[i,:]
        for j in range(X.shape[0]):
            x2 = X[j,:]
            if (G[i,j] == 1):
                x1_values = [x1[0], x2[0]]
                x2_values = [x1[1], x2[1]]
                if (Y[i]!=Y[j]):
                    if [i,j] not in SE and [j,i] not in SE:
                        SE.append([i,j])
                    if(plot):
                            plt.plot(x1_values, x2_values, 'r')
                elif(plot):
                    plt.plot(x1_values, x2_values, 'k')
    return SE

SE = getSupportEdges(X, Y, G, True)

print('Vertex indices on each support edge' + str(SE))
```

```python
def removeOverlapping(G, X, Y, plot=False):

    D_x = np.sum(G, axis = 1 )
    D_x_classe1 = np.sum(G[0:int(N1),0:int(N1)], axis = 1 )
    D_x_classe2 = np.sum(G[int(N1):G.shape[0],int(N1):G.shape[0]], axis = 1 )
    D_x_classe = np.hstack((D_x_classe1,D_x_classe2))

    q = D_x_classe/ D_x

    Q_1 = q[0:D_x_classe1.shape[0],]
    Q_2 = q[D_x_classe1.shape[0]:G.shape[0],]
    t_1 = np.sum(Q_1)/Q_1.shape[0]
    t_2 = np.sum(Q_2)/Q_2.shape[0]

    indexesSV1 = np.where(q >= t_1)
    indexesSV2 = np.where(q >= t_2)
    
    X_NoOverlapping = X[np.intersect1d(indexesSV1, indexesSV2),:]
    Y_NoOverlapping = Y[np.intersect1d(indexesSV1, indexesSV2),]

    G_NoOverlapping = GabrielGraph(X_NoOverlapping, Y_NoOverlapping, plot)
    return G_NoOverlapping, X_NoOverlapping, Y_NoOverlapping
```

```python
G_NoOverlapping, X_NoOverlapping, Y_NoOverlapping = removeOverlapping(G, X, Y, False)

SE_NoOverLapping = getSupportEdges(X_NoOverlapping, Y_NoOverlapping, G_NoOverlapping, True)

print('Vertex indices on each edge' + str(SE_NoOverLapping))
```

```python
def getSigma(SE_NoOverLapping, X_NoOverlapping):
    m = np.zeros((len(SE_NoOverLapping), X_NoOverlapping.shape[1]))
    i = 0
    sigma = []
    for edge in SE_NoOverLapping:
        m[i,:] = 0.5 * (X_NoOverlapping[edge[0],:] + X_NoOverlapping[edge[1],:])
        R = np.linalg.norm(X_NoOverlapping[edge[0]] - m[i,:]) 
        sigma.append([[R/3, 0],[0 ,  R/3]])
        i = i+1
    return sigma
```

```python
def pNEstimativa(x, mu, cov):
    p = 1/np.sqrt((2*np.pi) ** mu.shape[0] * np.linalg.det(cov)) * np.exp(-0.5*(x - mu).transpose().dot(np.linalg.pinv(cov)).dot(x-mu))
    return p
```

```python
x = np.arange(-3, 3, 0.1)
y = np.arange(-3, 3, 0.1)

dens1 = np.zeros((len(SE_NoOverLapping), x.shape[0], y.shape[0]))
dens2 = np.zeros((len(SE_NoOverLapping), x.shape[0], y.shape[0]))

sigma = getSigma(SE_NoOverLapping, X_NoOverlapping) 

for i in range(x.shape[0]):
    for j in range(y.shape[0]):
        for p in range(len(SE_NoOverLapping)):                   

            dens1[p,j,i] = pNEstimativa([x[i], y[j]], X_NoOverlapping[SE_NoOverLapping[p][0],:], sigma[p]) 
            dens2[p,j,i] = pNEstimativa([x[i], y[j]], X_NoOverlapping[SE_NoOverLapping[p][1],:], sigma[p]) 

fig, ax = plt.subplots(figsize=(6,6))

#for p in range(len(SE_NoOverLapping)):   
#    ax.contour(x,y, dens1[p,:,:],  cmap='RdGy')
#    ax.contour(x,y, dens2[p,:,:])

ax.contour(x,y, np.sum(dens1, axis=0),  cmap='RdGy')
ax.contour(x,y, np.sum(dens2, axis=0))

plotPoints(X_NoOverlapping, Y_NoOverlapping)

ax.contour(x,y,
       np.sum(dens1, axis = 0) - np.sum(dens2, axis = 0),
       levels=[0], cmap="Greys_r")


plt.plot([X_NoOverlapping[np.array(SE_NoOverLapping)[:,0],0], X_NoOverlapping[np.array(SE_NoOverLapping)[:,1],0]],
         [X_NoOverlapping[np.array(SE_NoOverLapping)[:,0],1], X_NoOverlapping[np.array(SE_NoOverLapping)[:,1],1]], '-k')
```

```python
def twospirals(n_points, noise=1):
    """
     Returns the two spirals dataset.
    """
    n = np.sqrt(np.random.rand(n_points,1)) * 780 * (2*np.pi)/360
    d1x = -np.cos(n)*n + np.random.rand(n_points,1) * noise
    d1y = np.sin(n)*n + np.random.rand(n_points,1) * noise
    return (np.vstack((np.hstack((d1x,d1y)),np.hstack((-d1x,-d1y)))), 
            np.hstack((np.zeros(n_points),np.ones(n_points))))

#Número de pontos em cada classe
N = 200

#Dados de Treinamento
X_trn, Y_trn = twospirals(N)
Y_trn[0:int(N)] = -1

#Sepera as classes únicas
classes = np.unique(Y_trn)

plt.title('training set')
plt.plot(X_trn[Y_trn==-1,0], X_trn[Y_trn==-1,1], '.', label='class 1')
plt.plot(X_trn[Y_trn==1,0], X_trn[Y_trn==1,1], '.', label='class 2')
plt.legend()
plt.show()
```
